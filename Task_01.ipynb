### df.unique() :

### 	df.nunique() :

### df.isnull() :

* The df.isnull() method in pandas is used to create a Boolean mask indicating where values are missing or null in a data frame. It returns a data frame of the same shape as the original data frame, where each element is either True if the corresponding value is null or False if it is not null.



```python
df.isnull()
```

### df.fillna() :

* The df.fillna() method in pandas is used to fill missing or null values in a data frame with specified values. It returns a new data frame with the missing values replaced by the provided values.

```python
df.fillna(2)
```

### df.sort_values() : 

* The df.sort_values() method in pandas is used to sort a data frame by one or more columns. It returns a new data frame with the rows sorted based on the specified column. You can also sort by multiple columns by passing a list of column names to the by parameter.

```python
df.sort_values("Quantity")
```

### df.value_counts() : 

* The df.value_counts() method in pandas is used to count the occurrences of unique values in a column or Series. It returns a new Series containing the count of each unique value, sorted in descending order by default.

##### Pandas Functions to Modify the Dataset... 

### df.copy() : 

* The df.copy() method in pandas is used to create a copy of a data frame. It returns a new data frame object that is a deep copy of the original data frame, including the data and the index. Creating a copy of a data frame can consume additional memory, so it's recommended to use it only when necessary, particularly for large datasets.

```python
df.copy()
```

### df.rename() : 

* The df.rename() method in pandas is used to change the names of columns or index labels in a data frame. It's important to note that df.rename() returns a new data frame with the updated names, leaving the original data frame unchanged. If you want to modify the data frame in place, you can set the inplace = True parameter.


```python
df.rename(columns= {"OrderID" : "Customer Id" , "OrderDate" : "Date"})
```

### df.where() : 

* In the context of datasets, the df.where() method is typically used in pandas, a popular data manipulation library in Python. The df.where() function allows you to selectively update values in a data frame based on a specified condition, while keeping the original structure of the data frame intact.


```python
condition = df["Quantity"] > 50
result = df.where(condition , other = 1 )
result
```

### df.drop() : 

* In the context of datasets, the df.where() method is typically used in pandas, a popular data manipulation library in Python. The df.where() function allows you to selectively update values in a data frame based on a specified condition, while keeping the original structure of the data frame intact.

```python
df.drop(['Quantity'], axis=1, inplace=True)
df
```
